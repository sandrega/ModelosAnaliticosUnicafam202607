##LISTAS

```python
# UNA LISTA VACIA
mi_lista=[]
# LISTA CON VARIOS ELEMENTOS
mi_lista=[1,2,3,"4",5.0, True]
print(mi_lista[2])
```

```python
# CAMBIOS EN LA LISTA
mi_lista[0]="saul andres garcia"
print(mi_lista)
```

```python
# AGREGAR DOCUMENTOS A UNA LISTA
mi_lista.append("garcia")
print(mi_lista)
```

```python
# AGREGAR ELEMENTOS A UNA POSICION ESPECIFICA
mi_lista.insert(3,"casabianca")
print(mi_lista)
```

```python
# AGREGAR VARIOS ELEMENTOS A LISTA
mi_lista.extend([7,False])
print(mi_lista)
```

```python
# QUITAR ELEMENTOS
# ELIMINAR OBJETOS ESPECIFICOS
mi_lista.remove(True)
print(mi_lista)
```

```python
print("ACTUALIZACION DE DATOS")
valor_buscar=input("por favor,ingrese el nombre para actualizar")
valor_reemplazar=input("ingrese valor a reeplzar")
mi_lista[mi_lista.index(valor_buscar)]=valor_reemplazar
print("cambio ejucutado.imprimiendo resultado")
print(mi_lista)
```

```python
# CREA UN PROGRAMA QUE DETERMINE SI UNA PERSONA ES MAYOR DE EDAD O MENOR DE EDAD

print("----- VERIFICACIÓN DE EDAD -----")

edad = int(input("Ingresa tu edad: "))

if edad < 18:
    print(f"Eres menor de edad, tienes: {edad}")
else:
    print(f"Eres mayor de edad, tienes: {edad}")
```

```python
# CONDICIONALES MÚLTIPLES

nota = 50

if nota >= 80:
    if nota >= 90:
        print("Excelente")
    elif nota >= 80 and nota < 90:
        print("Aceptable")
else:
    print("Insuficiente")
```
