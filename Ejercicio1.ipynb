```python
# EJERCICIO 4: OPERADORES MATEMÁTICOS

a = 10
b = 3
```

```python
# SUMA

suma = a + b

print(suma)
```

```python
# RESTA

resta = a - b

print(resta)
```

```python
# MULTIPLICACIÓN

multiplicacion = a * b

print(multiplicacion)
```

```python
# DIVISIÓN

division = a / b

print(division)
```

```python
# MÓDULO

modulo = a % b

print(modulo)
```

```python
# POTENCIA

potencia = a ** b

print(potencia)
```

```python
# DIVISIÓN ENTERA

division_entera = a // b

print(division_entera)
```

```python
# EJERCICIO 5: VALORES DE CADENA

saludo = "Hola, mundo!"
```

```python
# PRIMER CARÁCTER

print(saludo[0])
```

```python
# ÚLTIMOS 5 CARACTERES

print(saludo[-5:])
```

```python
# CONCATENAR CADENAS

cadena2 = ", ¿cómo estás?"

saludo_completo = saludo + " " + cadena2

print(saludo_completo)
```

```python
# MAYÚSCULAS

saludo_mayusculas = saludo_completo.upper()

print(saludo_mayusculas)
```

```python
# REEMPLAZAR TEXTO

saludo_reemplazo = saludo_completo.replace("mundo", "Python")

print(saludo_reemplazo)
```
