```python
# DICCIONARIO DEL PERFIL

perfil = {
    "nombre": "Saul Andres Garcia Montoya",
    "edad": 28,
    "ciudad": "Bogotá",
    "correo": "sandrega2211@gmail.com",
    "carrera": "Analítica de Datos",
    "habilidades": ["Python", "Excel", "Power BI"]
}

print(perfil)
```

```python
# MENÚ DE OPCIONES

print("========== MENÚ ==========")
print("1. Agregar un dato")
print("2. Actualizar un dato")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con pop")
print("5. Consultar un dato")
print("6. Ver perfil completo")

opcion = input("Seleccione una opción: ")
```

```python
# OPCIÓN 1: AGREGAR UN DATO NUEVO

if opcion == "1":
    clave = input("Ingrese la nueva clave: ")
    valor = input("Ingrese el valor: ")

    perfil[clave] = valor

    print("Dato agregado correctamente.")
    print(perfil)
```

```python
# OPCIÓN 2: ACTUALIZAR UN DATO EXISTENTE

if opcion == "2":
    clave = input("Ingrese la clave que desea actualizar: ")

    if clave in perfil:
        nuevo_valor = input("Ingrese el nuevo valor: ")
        perfil[clave] = nuevo_valor

        print("Dato actualizado correctamente.")
        print(perfil)
    else:
        print("La clave no existe.")
```

```python
# OPCIÓN 3: ELIMINAR UN DATO CON DEL

if opcion == "3":
    clave = input("Ingrese la clave que desea eliminar: ")

    if clave in perfil:
        del perfil[clave]
        print("Dato eliminado correctamente.")
        print(perfil)
    else:
        print("La clave no existe.")
```

```python
# OPCIÓN 4: ELIMINAR UN DATO CON POP

if opcion == "4":
    clave = input("Ingrese la clave que desea eliminar: ")

    if clave in perfil:
        perfil.pop(clave)
        print("Dato eliminado correctamente.")
        print(perfil)
    else:
        print("La clave no existe.")
```

```python
# OPCIÓN 5: CONSULTAR UN DATO

if opcion == "5":
    clave = input("Ingrese la clave que desea consultar: ")

    if clave in perfil:
        print("Valor encontrado:")
        print(perfil[clave])
    else:
        print("La clave no existe.")
```

```python
# OPCIÓN 6: VER PERFIL COMPLETO

if opcion == "6":
    print("===== PERFIL COMPLETO =====")

    for clave, valor in perfil.items():
        print(clave, ":", valor)
```
