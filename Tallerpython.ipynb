```python
# Ejercicio 2: Clasificador de IMC

peso = 70
altura = 1.75

imc = peso / altura ** 2

print(f"IMC: {imc:.1f}")

if imc < 18.5:
    print("Bajo peso")
elif imc < 25:
    print("Normal")
elif imc < 30:
    print("Sobrepeso")
else:
    print("Obesidad")
```

```python
# Ejercicio 4: Acceso a sistema

edad = 20
password = "abc123"

mayor = edad > 18
pass_ok = len(password) > 8

if mayor and pass_ok:
    print("Acceso permitido")
elif not mayor:
    print("Acceso denegado: edad insuficiente")
elif not pass_ok:
    print("Acceso denegado: contraseña muy corta")
```

```python
# Ejercicio 5: Calculadora de descuento

monto = 120000

if monto < 50000:
    descuento = 0
elif monto < 100000:
    descuento = 0.10
elif monto < 200000:
    descuento = 0.20
else:
    descuento = 0.30

ahorro = monto * descuento
total = monto - ahorro

print(f"Descuento: {int(descuento * 100)}% (${ahorro:,.0f})")
print(f"Total: ${total:,.0f}")
```

```python
# Ejercicio 6: FizzBuzz

def fizzbuzz(n):
    if n % 3 == 0 and n % 5 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)

fizzbuzz(15)
fizzbuzz(9)
fizzbuzz(10)
fizzbuzz(7)
```

```python
# Ejercicio 2: Suma acumulada

total = 0

for i in range(1, 101):
    total = total + i

print("La suma del 1 al 100 es:", total)

pares = 0

for i in range(2, 101, 2):
    pares = pares + i

print("La suma de los números pares es:", pares)
```

```python
# Ejercicio 3: Recorrer una lista

precios = [12000, 45000, 8500, 67000, 3200]

caros = 0

for precio in precios:
    print(precio)

    if precio > 10000:
        caros = caros + 1

print("Cantidad de artículos mayores a $10.000:", caros)
```

```python
# Ejercicio 4: While con condición

saldo = 1000
mes = 0

while saldo > 0:
    saldo = saldo - 150
    saldo = saldo - 20
    mes = mes + 1
    print("Mes", mes, "- Saldo:", saldo)

print("El saldo se agotó en", mes, "meses")
```

```python
# Ejercicio 5: break y continue

numeros = [3, 7, -2, 15, 4, -8, 9, 1]

for numero in numeros:
    if numero < 0:
        continue

    if numero > 10:
        print("Encontré", numero, ", deteniendo.")
        break

    print("Procesando:", numero)
```
